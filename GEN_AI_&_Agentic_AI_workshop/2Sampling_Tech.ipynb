```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
```

```python
# Load pre-trained model and tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
prompt = "The future of AI is"
input_ids = tokenizer.encode(prompt,
                             return_tensors='pt')
```

```python
# Greedy
print("Greedy:")
print(tokenizer.decode(model.generate(input_ids,
                                      max_length=50,
                                      do_sample=False)[0]))
```

```python
# Beam Search
print("Beam Search:")
print(tokenizer.decode(model.generate(input_ids,
                                      max_length=50,
                                      num_beams=5,
                                      early_stopping=True)[0]))
```

```python
# Top-K
print("Top-K Sampling:")
print(tokenizer.decode(model.generate(input_ids,
                                      do_sample=True,
                                      top_k=50,
                                      max_length=50)[0]))
```

```python
# Top-P
print("Top-P Sampling:")
print(tokenizer.decode(model.generate(input_ids,
                                      do_sample=True,
                                      top_p=0.7,
                                      max_length=50)[0]))
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-medium")
model = GPT2LMHeadModel.from_pretrained("gpt2-medium")

"""model.eval()
torch.no_grad()""";
```

```python
# greedy
# Always picks the word with the highest probability at each step. No randomness.
# Prompt to generate text from
prompt_text = "In the future, artificial intelligence will"

# Tokenize input
input_ids = tokenizer.encode(prompt_text, return_tensors='pt')

# Generate text using greedy decoding (no sampling)
output_ids = model.generate(
    input_ids=input_ids,
    max_length=50,         # Total length (input + generated)
    do_sample=False        # No randomness; always pick the best next token
)

# Decode and print result
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(" Greedy Decoding Output:\n", output_text)
```

```python
# Beam Search
# Explores multiple paths simultaneously and chooses the best overall sequence.
beam_output = model.generate(
    input_ids=input_ids,
    max_length=50,
    num_beams=5,             # Explore 5 possible sequences at each step
    no_repeat_ngram_size=2,  # Prevent repeating phrases
    early_stopping=True      # Stop if all beams reach the end token
)

output_text = tokenizer.decode(beam_output[0],
                               skip_special_tokens=True)
print("Beam Search Output:\n", output_text)
```

```python
# Top-K Sampling generation
top_k_output = model.generate(
    input_ids=input_ids,
    do_sample=True,      # Enable randomness/sampling
    top_k=50,            # Sample from top 50 tokens only
    max_length=50
)

output_text = tokenizer.decode(top_k_output[0], skip_special_tokens=True)
print("🎲 Top-K Sampling Output:\n", output_text)
```

```python
# Top-P Sampling (Nucleus Sampling)
top_p_output = model.generate(
    input_ids=input_ids,
    do_sample=True,      # Enable sampling
    top_p=0.9,           # Cumulative probability threshold
    top_k=0,             # Disable top-k to purely use nucleus sampling
    max_length=50
)

output_text = tokenizer.decode(top_p_output[0], skip_special_tokens=True)
print("🔮 Top-P (Nucleus) Sampling Output:\n", output_text)
```

```python
# combination of top p, top k and temperature

combo_output = model.generate(
    input_ids=input_ids,
    do_sample=True,
    top_k=50,
    top_p=0.92,
    temperature=0.8,     # Lower = safer, Higher = wider
    max_length=50
)

output_text = tokenizer.decode(combo_output[0],
                               skip_special_tokens=True)
print("🧪 Mixed Sampling (Top-K + Top-P + Temp=0.8):\n",
      output_text)
```
